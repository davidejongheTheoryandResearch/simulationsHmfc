```python
import pandas as pd

# Load files
symptom_df = pd.read_csv("Data/SymptomData.csv")
perceptual_df = pd.read_csv("Data/PerceptualTaskData.csv")
```

```python


id_column = "IDcode"


# Extract unique IDs from both datasets
symptom_ids = set(symptom_df[id_column].unique())
perceptual_ids = set(perceptual_df[id_column].unique())

# Print number of unique IDs
print(f"Number of unique IDs in SymptomData: {len(symptom_ids)}")
print(f"Number of unique IDs in PerceptualTaskData: {len(perceptual_ids)}")

# Check whether all SymptomData IDs are present in PerceptualTaskData
missing_ids = symptom_ids - perceptual_ids

if len(missing_ids) == 0:
    print("All unique IDs from SymptomData are present in PerceptualTaskData.")
else:
    print(f"{len(missing_ids)} ID(s) from SymptomData are missing in PerceptualTaskData.")
    print("Missing IDs:")
    print(missing_ids)

# Optional: also check the reverse (IDs in PerceptualTaskData not in SymptomData)
extra_ids = perceptual_ids - symptom_ids
if len(extra_ids) > 0:
    print(f"{len(extra_ids)} ID(s) are present in PerceptualTaskData but not in SymptomData.")
```
